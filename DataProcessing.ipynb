```
"""
这个文本是用来将视频中的音频转换为Wav格式。我们把该脚本放在data环境下。会自动的将一个文件夹的视频内容转换成另外一个文件夹内的音频内容。
注意在使用之前需要安装音视频处理库。
这个脚本将音频信号转换成单通道。44100采样率。每个音频文件每个文具时间为4秒。
音频信号与有声道采样率采样位数要么格式比特率。其中我们通过声道还有采样率以及采样位数来计算比特率。样本格式如mp3格式是属于一种头文件的形式。
"""

import os
import re
import subprocess

INPUT_VIDEO = "./baby_avi/"
OUTPUT_FILE = "./baby_wav/"
def replcaeFileName(pic_path):#修改pic_path路径下的文件名
    piclist=os.listdir(pic_path)
    total_num=len(piclist)
    i= 0
    for pic in piclist:
        if pic.endswith(".avi") :#修改成你自己想要重命名的文件格式
          old_path=INPUT_VIDEO+pic
          new_path=OUTPUT_FILE+str(1000+(int(i)))+'.wav' 
          print(old_path) 
          if os.path.isfile(new_path) is True:
            os.remove(new_path)      
          print(old_path)
          print(new_path)
          #cmd = "ffmpeg -i " + old_path + " -ab 160k -ac 2 -ar 44100 -vn " + new_path
          cmd = "ffmpeg -i "+ old_path +" -ac 1 -ar 44100 "+new_path
          #print(cmd)
          subprocess.call(cmd, shell=True)
      
          i=i+1




if __name__ == '__main__':
   
  replcaeFileName("./baby_avi/")
```
